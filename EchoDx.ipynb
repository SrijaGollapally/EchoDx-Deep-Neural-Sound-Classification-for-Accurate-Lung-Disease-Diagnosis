```python
# Install dependencies
!pip install --upgrade librosa
!pip install -q resampy
!pip install pywavelets

# Import libraries
import os
import zipfile
import wave
import random
import numpy as np
import pandas as pd
import seaborn as sns
import tensorflow as tf
import matplotlib.pyplot as plt
import pywt
import librosa
import librosa.display
from IPython.display import Audio
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import (
    Conv1D, MaxPooling1D, BatchNormalization, Input, Dense, Dropout,
    LSTM, Bidirectional, LeakyReLU, Add, Attention, GRU, Activation
)
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.regularizers import l2
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    matthews_corrcoef, cohen_kappa_score, roc_auc_score,
    confusion_matrix, classification_report
)
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.utils.class_weight import compute_class_weight
from google.colab import drive

print("Dependencies installed and libraries imported successfully!")

# Mount Google Drive
drive.mount('/content/drive')

# Extract training data
zip_path = "/content/drive/MyDrive/respiratory sound database.zip"
extract_path = "/content/respiratory_sound_data"
os.makedirs(extract_path, exist_ok=True)
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)
print("Training data extraction completed!")

# Define paths
audio_data = '/content/respiratory_sound_data/Respiratory_Sound_Database/Respiratory_Sound_Database/audio_and_txt_files/'
diagnosis_file = '/content/respiratory_sound_data/respiratory_sound_database/Respiratory_Sound_Database/patient_diagnosis.csv'

# Load diagnosis data
diagnosis_df = pd.read_csv(diagnosis_file, names=['patient_id', 'disease'])
print("Diagnosis Data Head:")
print(diagnosis_df.head(4))
print("\nDisease Counts:")
print(diagnosis_df.disease.value_counts())

# Augmentation functions
def add_noise(data, noise_level=0.005):
    noise = np.random.randn(len(data))
    return data + noise_level * noise

def shift_audio(data, shift_max=2400):
    shift = random.randint(-shift_max, shift_max)
    return np.roll(data, shift)

def stretch_audio(data, rate=1.1):
    return librosa.effects.time_stretch(data, rate=rate)

def pitch_shift_audio(data, sr, n_steps=2):
    return librosa.effects.pitch_shift(data, sr=sr, n_steps=n_steps)

print("Augmentation functions defined!")

# Visualization functions
def show_audio(audio_path):
    y, sr = librosa.load(audio_path)
    y_noise = add_noise(y, 0.005)
    y_shift = shift_audio(y, 2400)
    y_stretch_1 = stretch_audio(y, 1.1)
    y_pitch_shift = pitch_shift_audio(y, sr, 2)

    plt.figure(figsize=(20, 8))
    plt.subplot(2, 2, 1)
    librosa.display.waveshow(y, sr=sr)
    plt.title('Original')
    plt.subplot(2, 2, 2)
    librosa.display.waveshow(y_noise, sr=sr)
    plt.title('Noise')
    plt.subplot(2, 2, 3)
    librosa.display.waveshow(y_shift, sr=sr)
    plt.title('Shift')
    plt.subplot(2, 2, 4)
    librosa.display.waveshow(y_stretch_1, sr=sr)
    plt.title('Stretch 1.1')
    plt.tight_layout()
    plt.show()

def show_audio_features(audio_path):
    y, sr = librosa.load(audio_path)
    y_noise = add_noise(y, 0.005)
    y_shift = shift_audio(y, 2400)
    y_stretch_1 = stretch_audio(y, 1.1)
    y_pitch_shift = pitch_shift_audio(y, sr, 2)

    mfcc_orig = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=60)
    mfcc_noise = librosa.feature.mfcc(y=y_noise, sr=sr, n_mfcc=60)
    mfcc_shift = librosa.feature.mfcc(y=y_shift, sr=sr, n_mfcc=60)
    mfcc_stretch_1 = librosa.feature.mfcc(y=y_stretch_1, sr=sr, n_mfcc=60)
    mfcc_pitch_shift = librosa.feature.mfcc(y=y_pitch_shift, sr=sr, n_mfcc=60)

    plt.figure(figsize=(20, 8))
    plt.subplot(2, 2, 1)
    librosa.display.specshow(librosa.power_to_db(mfcc_orig, ref=np.max), y_axis='mel', fmax=8000, x_axis='time')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Original')
    plt.subplot(2, 2, 2)
    librosa.display.specshow(librosa.power_to_db(mfcc_noise, ref=np.max), y_axis='mel', fmax=8000, x_axis='time')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Noise')
    plt.subplot(2, 2, 3)
    librosa.display.specshow(librosa.power_to_db(mfcc_shift, ref=np.max), y_axis='mel', fmax=8000, x_axis='time')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Shift')
    plt.subplot(2, 2, 4)
    librosa.display.specshow(librosa.power_to_db(mfcc_stretch_1, ref=np.max), y_axis='mel', fmax=8000, x_axis='time')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Stretch 1.1')
    plt.tight_layout()
    plt.show()

print("Visualization functions defined!")

# Show sample visualizations
show_audio(os.path.join(audio_data, '104_1b1_Lr_sc_Litt3200.wav'))
show_audio_features(os.path.join(audio_data, '104_1b1_Lr_sc_Litt3200.wav'))

# Wavelet feature extraction
def advanced_wavelet_features(audio, wavelet='db4', level=6, epsilon=1e-10):
    coeffs = pywt.wavedec(audio, wavelet=wavelet, level=level)
    feature_list = []
    for c in coeffs:
        if len(c) == 0:
            continue
        energy = np.sum(c**2)
        entropy = -np.sum(c * np.log2(np.abs(c) + epsilon))
        skewness = np.mean(((c - np.mean(c)) / (np.std(c) + epsilon))**3)
        kurtosis = np.mean(((c - np.mean(c)) / (np.std(c) + epsilon))**4) - 3
        feature_list.extend([energy, entropy, skewness, kurtosis])
    return np.array(feature_list, dtype=np.float32)

print("Wavelet feature extraction function defined!")

# Feature extraction with reduced augmentation
def mfccs_feature_extraction(dir_):
    X_ = []
    y_ = []
    COPD = []
    copd_count = 0
    data = diagnosis_df
    mfcc_dim = 60
    wavelet_type = 'db4'
    wavelet_level = 6

    for soundDir in os.listdir(dir_):
        if soundDir.endswith('.wav') and soundDir[:3] not in ['103', '108', '115']:
            pid = int(soundDir[:3])
            p = data[data['patient_id'] == pid]['disease'].values[0]
            if p in ['Bronchiectasis', 'Bronchiolitis']:
                p = 'Bronchiolitis'
            file_path = os.path.join(dir_, soundDir)
            data_x,sampling_rate = librosa.load(file_path, res_type='kaiser_fast')

            if p == 'COPD':
                if (soundDir[:6] in COPD) and copd_count < 2:
                    mfccs_ = np.mean(librosa.feature.mfcc(y=data_x, sr=sampling_rate, n_mfcc=mfcc_dim).T, axis=0)
                    wavelet_feats_ = advanced_wavelet_features(data_x, wavelet=wavelet_type, level=wavelet_level)
                    combined_feats_ = np.concatenate([mfccs_, wavelet_feats_])
                    COPD.append(soundDir[:6])
                    copd_count += 1
                    X_.append(combined_feats_)
                    y_.append(p)
                if (soundDir[:6] not in COPD):
                    mfccs_ = np.mean(librosa.feature.mfcc(y=data_x, sr=sampling_rate, n_mfcc=mfcc_dim).T, axis=0)
                    wavelet_feats_ = advanced_wavelet_features(data_x, wavelet=wavelet_type, level=wavelet_level)
                    combined_feats_ = np.concatenate([mfccs_, wavelet_feats_])
                    COPD.append(soundDir[:6])
                    copd_count = 0
                    X_.append(combined_feats_)
                    y_.append(p)
            else:
                # Original
                mfccs_ = np.mean(librosa.feature.mfcc(y=data_x, sr=sampling_rate, n_mfcc=mfcc_dim).T, axis=0)
                wavelet_feats_ = advanced_wavelet_features(data_x, wavelet=wavelet_type, level=wavelet_level)
                combined_feats_ = np.concatenate([mfccs_, wavelet_feats_])
                X_.append(combined_feats_)
                y_.append(p)
                # Noise
                data_noise = add_noise(data_x, 0.005)
                mfccs_noise = np.mean(librosa.feature.mfcc(y=data_noise, sr=sampling_rate, n_mfcc=mfcc_dim).T, axis=0)
                wavelet_noise = advanced_wavelet_features(data_noise, wavelet=wavelet_type, level=wavelet_level)
                combined_noise = np.concatenate([mfccs_noise, wavelet_noise])
                X_.append(combined_noise)
                y_.append(p)
                # Shift
                data_shift_ = shift_audio(data_x, 2400)
                mfccs_shift_ = np.mean(librosa.feature.mfcc(y=data_shift_, sr=sampling_rate, n_mfcc=mfcc_dim).T, axis=0)
                wavelet_shift_ = advanced_wavelet_features(data_shift_, wavelet=wavelet_type, level=wavelet_level)
                combined_shift_ = np.concatenate([mfccs_shift_, wavelet_shift_])
                X_.append(combined_shift_)
                y_.append(p)
                # Stretch
                data_stretch_ = stretch_audio(data_x, 1.1)
                mfccs_stretch_ = np.mean(librosa.feature.mfcc(y=data_stretch_, sr=sampling_rate, n_mfcc=mfcc_dim).T, axis=0)
                wavelet_stretch_ = advanced_wavelet_features(data_stretch_, wavelet=wavelet_type, level=wavelet_level)
                combined_stretch_ = np.concatenate([mfccs_stretch_, wavelet_stretch_])
                X_.append(combined_stretch_)
                y_.append(p)

    X_data = np.array(X_)
    y_data = np.array(y_)
    scaler = StandardScaler()
    X_data = scaler.fit_transform(X_data)
    return X_data, y_data

print("Feature extraction function defined!")

# Extract features
x_mfccs, y = mfccs_feature_extraction(audio_data)
print("Feature shape:", x_mfccs.shape)
print("Labels shape:", y.shape)

# Encode labels
label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(y)
onehot_encoder = OneHotEncoder(sparse_output=False)
y_data_encoded = onehot_encoder.fit_transform(integer_encoded.reshape(-1, 1))

mfccs_features = np.array(x_mfccs)
labels = np.array(y_data_encoded)
print("Final feature shape:", mfccs_features.shape)
print("Final label shape:", labels.shape)

# Adjusted data split with stratification
mfcc_temp, mfcc_test, labels_temp, labels_test = train_test_split(
    mfccs_features, labels, test_size=0.15, random_state=42, stratify=labels
)
mfcc_train, mfcc_val, labels_train, labels_val = train_test_split(
    mfcc_temp, labels_temp, test_size=0.1765, random_state=42, stratify=labels_temp
)
print("Train Labels Shape:", labels_train.shape)
print("Validation Labels Shape:", labels_val.shape)
print("Test Labels Shape:", labels_test.shape)

# Prepare data for model
x_train_gru = np.expand_dims(mfcc_train, axis=1)
x_val_gru = np.expand_dims(mfcc_val, axis=1)
x_test_gru = np.expand_dims(mfcc_test, axis=1)
y_train_gru = labels_train
y_val_gru = labels_val
y_test_gru = labels_test

# Define model with adjusted capacity
input_layer = Input(shape=(1, x_mfccs.shape[1]))
conv1 = Conv1D(96, kernel_size=5, padding='same', activation='relu', kernel_regularizer=l2(0.02))(input_layer)  # Increased filters
conv1 = MaxPooling1D(pool_size=2, padding='same')(conv1)
conv1 = BatchNormalization()(conv1)
conv2 = Conv1D(192, kernel_size=5, padding='same', activation='relu', kernel_regularizer=l2(0.02))(conv1)  # Increased filters
conv2 = MaxPooling1D(pool_size=2, padding='same')(conv2)
conv2 = BatchNormalization()(conv2)
conv3 = Conv1D(192, kernel_size=5, padding='same', activation='relu', kernel_regularizer=l2(0.02))(conv2)
conv3 = MaxPooling1D(pool_size=2, padding='same')(conv3)
conv3 = BatchNormalization()(conv3)
conv3_adjusted = Conv1D(96, kernel_size=1, padding='same')(conv3)
residual = Add()([conv1, conv3_adjusted])
bi_lstm = Bidirectional(LSTM(192, return_sequences=True))(residual)  # Increased units
attention = Attention()([bi_lstm, bi_lstm])
bi_lstm = Bidirectional(LSTM(48, return_sequences=False))(attention)  # Increased units
fc1 = Dense(48, kernel_regularizer=l2(0.02))(bi_lstm)  # Increased units
fc1 = LeakyReLU()(fc1)
fc1 = Dropout(0.5)(fc1)  # Reduced dropout
fc2 = Dense(24, kernel_regularizer=l2(0.02))(fc1)  # Increased units
fc2 = LeakyReLU()(fc2)
fc2 = Dropout(0.5)(fc2)
output_layer = Dense(5, activation='softmax')(fc2)
cnn_lstm_model = Model(inputs=input_layer, outputs=output_layer)

# Compile model with adjusted learning rate
initial_lr = 0.0003
optimiser = Adam(learning_rate=initial_lr)
cnn_lstm_model.compile(
    optimizer=optimiser,
    loss=tf.keras.losses.CategoricalCrossentropy(label_smoothing=0.1),  # Reduced smoothing
    metrics=['accuracy']
)
cnn_lstm_model.summary()

# Compute class weights
y_integers = np.argmax(labels, axis=1)
class_weights = compute_class_weight('balanced', classes=np.unique(y_integers), y=y_integers)
class_weight_dict = dict(enumerate(class_weights))

# Define callbacks
cb = [
    EarlyStopping(patience=8, monitor='val_accuracy', mode='max', restore_best_weights=True),  # Tighter patience, max accuracy
    ModelCheckpoint("diagnosis_LSTM_CNN.h5", save_best_only=True, monitor='val_accuracy', mode='max'),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=4, min_lr=1e-6)  # Tighter patience
]

# Create datasets
train_dataset = tf.data.Dataset.from_tensor_slices((x_train_gru, y_train_gru)).shuffle(1024).batch(32).prefetch(tf.data.AUTOTUNE)
val_dataset = tf.data.Dataset.from_tensor_slices((x_val_gru, y_val_gru)).batch(32)
test_dataset = tf.data.Dataset.from_tensor_slices((x_test_gru, y_test_gru)).batch(32)

# Train model
history = cnn_lstm_model.fit(
    train_dataset,
    epochs=50,
    validation_data=val_dataset,
    callbacks=cb,
    class_weight=class_weight_dict
)

# Evaluate model
train_eval = cnn_lstm_model.evaluate(x_train_gru, y_train_gru, verbose=0)
val_eval = cnn_lstm_model.evaluate(x_val_gru, y_val_gru, verbose=0)
test_loss, test_accuracy = cnn_lstm_model.evaluate(x_test_gru, y_test_gru, batch_size=32)
print(f"Train Loss: {train_eval[0]:.4f}, Train Accuracy: {train_eval[1] * 100:.2f}%")
print(f"Validation Loss: {val_eval[0]:.4f}, Validation Accuracy: {val_eval[1] * 100:.2f}%")
print(f"Test Loss: {test_loss:.4f}, Test Accuracy: {test_accuracy * 100:.2f}%")

# Plot training curves
def plot_loss_curves(history):
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    accuracy = history.history['accuracy']
    val_accuracy = history.history['val_accuracy']
    epochs = range(len(loss))
    plt.figure()
    plt.grid()
    plt.plot(epochs, accuracy, label='training_accuracy')
    plt.plot(epochs, val_accuracy, label='val_accuracy')
    plt.title('Accuracy')
    plt.xlabel('Epochs')
    plt.legend()
    plt.show()

plot_loss_curves(history)

# Confusion matrix and classification report
classes = ["COPD", "Bronchiolitis", "Pneumonia", "URTI", "Healthy"]
preds = cnn_lstm_model.predict(x_test_gru)
classpreds = [np.argmax(t) for t in preds]
y_testclass = [np.argmax(t) for t in y_test_gru]
cm = confusion_matrix(y_testclass, classpreds)
plt.figure(figsize=(8, 6), dpi=80)
sns.heatmap(cm, cmap='Blues', annot=True, fmt='d', xticklabels=classes, yticklabels=classes)
plt.xlabel('Prediction')
plt.ylabel('Truth')
plt.show()
print(classification_report(y_testclass, classpreds, target_names=classes))

# Prediction function
def predict_single_audio(test_file, model, classes, mfcc_dim=60, wavelet_type='db4', wavelet_level=6):
    data_x, sr = librosa.load(test_file, res_type='kaiser_fast')
    mfccs_ = np.mean(librosa.feature.mfcc(y=data_x, sr=sr, n_mfcc=mfcc_dim).T, axis=0)
    wavelet_ = advanced_wavelet_features(data_x, wavelet=wavelet_type, level=wavelet_level)
    combined_ = np.concatenate([mfccs_, wavelet_])
    combined_ = np.expand_dims(combined_, axis=0)
    combined_ = np.expand_dims(combined_, axis=1)
    preds = model.predict(combined_)
    pred_class_idx = np.argmax(preds[0])
    pred_class = classes[pred_class_idx]
    confidence = float(np.max(preds[0]))
    print(f"Predicted: {pred_class}, Confidence: {confidence:.4f}")
    return pred_class, confidence

# Extract test data
test_zip_path = "/content/drive/MyDrive/respiratory test.zip"
test_extract_path = "/content/respiratory_test_extracted"
os.makedirs(test_extract_path, exist_ok=True)
with zipfile.ZipFile(test_zip_path, 'r') as zip_ref:
    zip_ref.extractall(test_extract_path)
print(f"Extraction complete! Files are extracted to: {test_extract_path}")

# Predict on sample test file
test_file = '/content/respiratory_test_extracted/107_2b5_Ar_mc_AKGC417L.wav'
classes = ["COPD", "Bronchiolitis", "Pneumonia", "URTI", "Healthy"]
pred_class, confidence = predict_single_audio(test_file, cnn_lstm_model, classes)
```
